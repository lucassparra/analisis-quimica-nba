```python
import pandas as pd
import matplotlib.pyplot as plt
from funciones import extraer_datos
```

```python
datos = pd.DataFrame(extraer_datos('Advanced.csv'))
print(datos.head())
```

```python
datos.dtypes
```

```python
#Elimino las variables que aparentemente no aportan nada al análisis
datos = datos.drop(['lg', 'team', 'pos', 'age', 'gs'], axis=1, errors='ignore')

#A continuación eliminamos estadísticas del jugador que miden la calidad del jugador
#pero no su estilo de juego que es lo que nos interesa.
datos = datos.drop(['per', 'ts_percent', 'ows', 'dws', 'ws', 'ws_48','obpm', 'dbpm', 'bpm', 'vorp'], axis=1, errors='ignore')

#Visualizamos las estadisticas restantes y sus tipos
print(datos.columns, datos.dtypes)
```

```python
#Las estadísticas de cada jugador están repartidas en varias filas (una por temporada)
#Por ejemplo, para Ricky Rubio:
print(f'Ricky Rubio aparece {(datos['player'] ==  'Ricky Rubio').sum()} veces en la base de datos.') #Aparece 12 veces

#Vamos a agrupar las filas por jugador, calculando la media de sus estadísticas
datos_agrupados = datos.groupby(['player_id', 'player']).mean().reset_index()
print(f'Ricky Rubio aparece {(datos_agrupados['player'] ==  'Ricky Rubio').sum()} veces en la base de datos agrupada por temporadas.') #Aparece 1 vez
```
